```python
import seaborn as sns
from pycaret.classification import *

penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
print(penguins.shape)
penguins.head()
```

##  Experiment 1 — Flipper & Bill Length → Species

```python
df1 = penguins[['flipper_length_mm', 'bill_length_mm', 'species']]
df1.head()
```

```python
s1 = setup(data=df1, target='species', session_id=42)
```

```python
best1 = compare_models()
```

```python
evaluate_model(best1)
```

```python
pred1 = predict_model(best1, data=df1)
pred1.head()
```

```python
save_model(best1, 'experiment1_model')
```
