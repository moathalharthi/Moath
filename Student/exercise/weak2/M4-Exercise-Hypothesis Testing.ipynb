```python
import seaborn as sns 
from scipy import stats
from scipy.stats import chi2_contingency
import pandas as pd
import matplotlib.pyplot as plt

```

```python
tips = sns.load_dataset("tips")
```

```python
#Hypothesis H1​: smoking is associated with time of visit

tips = sns.load_dataset("tips")

contingency_table = pd.crosstab(tips['smoker'], tips['time'])

chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"--- Hypothesis 1 Results ---")
print(f"Chi-Square Statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject Null Hypothesis (Smoking and Time are associated)")
else:
    print("Result: Fail to Reject Null Hypothesis (Smoking and Time are independent)")
```

```python
#Hypothesis H2​: the bigger the group the higher the tip

# Calculate Pearson correlation coefficient
correlation = tips['size'].corr(tips['tip'])

print("--- Hypothesis 2 Results ---")
print(f"Correlation between Group Size and Tip: {correlation:.4f}")



if correlation > 0.5:
    print("Result: Strong positive correlation.")
elif correlation > 0:
    print("Result: Weak positive correlation.")
else:
    print("Result: No positive correlation found.")
```

```python
#Hypothesis H3​: group size is different based on the time of visit
lunch_size = tips[tips['time'] == 'Lunch']['size']
dinner_size = tips[tips['time'] == 'Dinner']['size']

t_stat, p_value = stats.ttest_ind(lunch_size, dinner_size)

print("--- Hypothesis 3 Results ---")
print(f"Mean Group Size (Lunch): {lunch_size.mean():.2f}")
print(f"Mean Group Size (Dinner): {dinner_size.mean():.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject Null Hypothesis (Group size is significantly different between Lunch and Dinner)")
else:
    print("Result: Fail to Reject Null Hypothesis (No significant difference in group size)")
```

```python
#Hypothesis H4​: (… difference in tips based on the time of visit…)

lunch_tips = tips[tips['time'] == 'Lunch']['tip']
dinner_tips = tips[tips['time'] == 'Dinner']['tip']

t_stat, p_value = stats.ttest_ind(lunch_tips, dinner_tips)

print(f"--- Hypothesis 4 Results ---")
print(f"Average Tip (Lunch): ${lunch_tips.mean():.2f}")
print(f"Average Tip (Dinner): ${dinner_tips.mean():.2f}")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject Null Hypothesis (There is a significant difference in tips between Lunch and Dinner)")
else:
    print("Result: Fail to Reject Null Hypothesis (The difference in tips is not statistically significant)")
```
